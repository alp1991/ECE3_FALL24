##3. 3D Vector display with `pyplot.quiver(), Axes3D 3d projection`

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


v1 = np.array([5, 6, 2])
v2 = np.array([3, 0, 5])
print(np.dot(v1, v2))
v3 = np.array([-4, 3, 1])
print(np.dot(v2, v3))
v4 = np.array([0, 0, 0])
print(np.dot(v3, v4))
v5 = np.array([2, 3, -1])
print(np.dot(v3, v5)) # v3 and v5 are orthogonal

#v1 = np.array([5, 6])
#v2 = np.array([3, 0])
#v3 = np.array([-4, 3])
#v4 = np.array([0, 0])
#v5 = np.array([2, 3])
#fig, ax = plt.subplots(figsize=(10,10))
#origin=[0,0]

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
origin = [0, 0, 0]

ax.quiver(*origin, *v1, color='r', label='v1')
#ax.quiver(*origin, *v1, color='r', scale=1, scale_units='xyz', angles='xyz', label='v1')
ax.quiver(*origin, *v2, color='g', label='v2')
ax.quiver(*origin, *v3, color='b', label='v3')
#ax.quiver(*origin, *v4, color='y', label='v4')
ax.quiver(*origin, *v5, color='m', label='v5')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

###4.a Matrix Inverse and Multiplication

```python
A = np.random.randint(1,10,size=(3,3))
Ainv = np.linalg.inv(A)
#AinvA = np.multiply(A,Ainv) # Wrong!
AinvA = np.matmul(Ainv,A)
print("A:\n",A)
print("Ainv:\n",Ainv)
print("AinvA:\n",AinvA)
print("AinvA rounded:\n",np.round(AinvA))
b = np.random.randint(1,10,size=(3,1))
print("A:\n",A)
print("b:\n",b)
```

###4.c Displaying 3D vectors with `pyplot.quiver()`

```python
# create a 3D vector from random integers
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
v3d = np.random.randint(1, 10, size=3)
print(v3d)
# create a diagonal matrix from that vector
D = np.diag(v3d)
print("Diagonal matrix:\n",D)
# compute its inverse
Dinv = np.linalg.inv(D)
print("Dinv:\n",Dinv)
# create a 2x3 matrix
v23 = np.random.randint(-10,10,size=(2,3))
print("v2x3:\n",v23)
# attempt to compute its inverse
#M23inv = np.linalg.inv(M23) # produces LinAlgError
#print(M23inv)
# plot v3d
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
origin = [0, 0, 0]
ax.quiver(*origin, *v3d, color='r', label='v3d')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.set_title("Plot of v3d")
ax.legend()
plt.tight_layout()
plt.show()

```
